# Delivery Platform Preference Analysis

The report is a synthesis of key insights from customer feedback on the subject of
online food delivery services in Taiwan, concentrating on frequency of use, the most
popular platforms and other aspects that matter when people make choices.

The findings presented in this report are based on a survey conducted via Google
Form (https://forms.gle/LdnWyiDtpTq4yU8s8) and distributed through social media, receiving responses from 28
participants. Although the number of responses is relatively small, they provide
valuable insights into the preferences and experiences of users within Taiwan's
online food delivery market

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency

# Read CSV and Inspect
data = pd.read_csv('survey_responses.csv')
print("Data Inspection:")
print(data.head())
print(data.info())

print("\nData Exploration:")
print("Usage Frequency Distribution:")
print(data['Frequency'].value_counts())
print("\nPreferred Platform Distribution:")
print(data['Platform'].value_counts())

# Data Visualization
plt.figure(figsize=(10, 6))
data['Frequency'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Usage Frequency Distribution')
plt.xlabel('Frequency')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(10, 6))
data['Platform'].value_counts().plot(kind='bar', color='salmon')
plt.title('Preferred Platform Distribution')
plt.xlabel('Platform')
plt.ylabel('Count')
plt.show()
```

```python
# Create a contingency table
contingency_table = pd.crosstab(data['Frequency'], data['Platform'])

# Perform chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)
print("\nChi-square test results:")
print("Chi-square statistic:", chi2)
print("p-value:", p)

# Interpretation of chi-square test:
if p < 0.05:
    print("\nThere is a significant association between Usage Frequency and Preferred Platform.")
else:
    print("\nThere is no significant association between Usage Frequency and Preferred Platform.")
```
